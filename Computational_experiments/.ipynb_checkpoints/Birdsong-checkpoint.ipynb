```python
import numpy              as np
import matplotlib.pyplot  as plt
from mpl_toolkits.mplot3d import Axes3D
from ripser               import Rips
```

```python
s = np.loadtxt('Presion_canario.txt')
```

```python
plt.figure(figsize=(25,4))
plt.plot(s, color = 'lightcoral')
plt.savefig('canario_original_time_series.png',bbox_inches='tight')
```

```python
plt.figure(figsize=(25,4))
plt.plot(range(0,115000), s[0:115000], color = 'coral')
plt.plot(range(115000,180000),s[115000:180000], color = 'darkred')
plt.plot(range(180000,230000),s[180000:230000], color = 'darkblue')
plt.plot(range(230000,len(s)),s[230000:len(s)], color = 'grey')
plt.savefig('canario_time_series.png',  bbox_inches='tight')
```

```python
n=0
N=len(s)
T = 500
step = 1
```

```python
xs = s[n:N-2*T:step]
ys = s[n+T:N-T:step]
zs = s[n+2*T:N:step]
```

```python
embedding = np.column_stack((xs, ys, zs))
```

```python
len(embedding)
```

```python
fig = plt.figure(figsize=(15,10))
ax = Axes3D(fig)

ax.scatter(xs, ys, zs, color = 'lightcoral', s=0.5, alpha = 0.2)
plt.tick_params(labelsize=15)
ax.set_title('Delay embedding')

plt.savefig('canario_embedding_scatter.png',  bbox_inches='tight')
```

```python
cuts= [0,115000, 180000, 230000, len(s)]
```

```python
T = 500
step = 1
```

```python
fig = plt.figure(figsize=(50,9))
color = ['coral', 'darkred', 'darkblue', 'grey']
subplot = [111,112,113,114]
for i in range(4):
    n=cuts[i]
    N=cuts[i+1]
    xs = s[n:N-2*T:step]
    ys = s[n+T:N-T:step]
    zs = s[n+2*T:N:step]
    ax = fig.add_subplot(1,4,i+1, projection = '3d')
    ax.plot(xs, ys, zs, lw=0.3, color = color[i])
    plt.tick_params(labelsize=10)

plt.savefig('canario_embedding_cut.png' ,  bbox_inches='tight')
```

```python
from scipy.spatial import  distance_matrix
from fermat import Fermat

def compute_fermat_distances(data, alpha, k):
    
    #Compute euclideandistances
    distances = distance_matrix(data,data)
    
    # Initialize the model
    fermat = Fermat(alpha = alpha, path_method='D', k = k) 

    # Fit
    fermat.fit(distances)
    
    ##Compute Fermat distances
    fermat_dist = fermat.get_distances()
    
    return  fermat_dist
```

```python
%%time
fermat_distances = compute_fermat_distances(embedding, 1.7, 100)
```

```python
from ripser import Rips
```

```python
rips = Rips()
diagrams2 = rips.fit_transform(fermat_distances, distance_matrix = True)
rips.plot(diagrams2, lifetime=False)
```
