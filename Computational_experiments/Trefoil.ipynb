## Persistence diagrams

### Euclidean Distance

```python
rips = Rips()
dgms = rips.fit_transform(data)
rips.plot(dgms, lifetime=True)
plt.title('Euclidean Distance')
```

### Isomap distance

```python
rips = Rips()
matrix = compute_adj_distance(data, 10)
dgms = rips.fit_transform(matrix, distance_matrix=True)
rips.plot(dgms, lifetime=True)
plt.title('Isomap distance k=10')
plt.show()
```

### Fermat Distance

```python
dgms = Fermat_dgm(data, p=3, rescaled=False)
```

```python
rips.plot([dgms[0][:0], dgms[1]], lifetime=True, title='Fermat Distance p=3 ($H_1$ only)')
plt.legend('')
plt.savefig('PH_trefoil_p=3_zoom.png',  bbox_inches='tight')
```
